```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
def read_img(path):
    with fits.open(path) as img:
    
        image = img[0].data
    return image
```

```python
p = 'D:/YASH/practice/NGC5548/x0l30801t_c0f.fits'
```

```python
image = read_img(p)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image, cmap='inferno')
plt.show()
```

```python
plt.axis('off')
plt.imshow(image[180:250,180:300], cmap='inferno')
plt.show()
```
